# Preguntas de negocio

##### 1) Puntualidad de los vuelos por aerolínea: ¿Cuál es el porcentaje de vuelos que salieron a tiempo (con un retraso de menos de 15 minutos) para cada aerolínea durante el año?

```python
# Filtrar vuelos con un retraso de menos de 15 minutos
flights_on_time = df_flights[df_flights['dep_delay'] <= 15]

# Contar el número de vuelos a tiempo por aerolínea
on_time_counts = flights_on_time.groupby('carrier').size()

# Contar el número total de vuelos por aerolínea
total_counts = df_flights.groupby('carrier').size()

# Calcular el porcentaje de vuelos a tiempo para cada aerolínea
punctuality_percentage = (on_time_counts / total_counts) * 100

# Mostrar el resultado
print(punctuality_percentage)
```

##### 2) Impacto del clima en los retrasos: ¿Existe una correlación entre el clima (temperatura, visibilidad, etc.) y los retrasos de los vuelos? ¿En qué medida afecta el clima a la puntualidad de las diferentes aerolíneas y rutas?

```python
# Combinar los datos de vuelos con los datos meteorológicos
merged_data = pd.merge(df_flights, df_weather, on=['origin', 'year', 'month', 'day', 'hour'])

# Calcular la correlación entre las variables meteorológicas y los retrasos de vuelos
correlation = merged_data[['dep_delay', 'arr_delay', 'temp', 'visib']].corr()

# Mostrar la correlación
print("Correlation Matrix:")
print(correlation)

# Calcular el impacto del clima en la puntualidad de las aerolíneas
impact_by_airline = merged_data.groupby('carrier')[['dep_delay', 'arr_delay', 'temp', 'visib']].mean()

# Mostrar el impacto del clima en la puntualidad de las aerolíneas
print("\nImpacto del clima en la puntualidad de las diferentes aerolíneas:")
print(impact_by_airline)
```

##### 3) ¿Cuáles son los aeropuertos más transitados en términos de número de vuelos?

```python
# Contar el número de vuelos por aeropuerto (tanto de origen como de destino)
departure_airports = df_flights['origin'].value_counts()
arrival_airports = df_flights['dest'].value_counts()

# Combinar los recuentos de vuelos de origen y destino para obtener el total de vuelos por aeropuerto
total_flights_per_airport = departure_airports.add(arrival_airports, fill_value=0)

# Ordenar los aeropuertos por el número de vuelos en orden descendente
top_airports = total_flights_per_airport.sort_values(ascending=False)

# Mostrar los aeropuertos más transitados
print("Los aeropuertos más transitados en términos de número de vuelos son:")
print(top_airports.head())
```

##### 6) ¿Hay alguna relación entre la edad de un avión y la frecuencia de mantenimiento requerido?

```python
# Calcular la edad de cada avión
from datetime import datetime
current_year = datetime.now().year
df_planes['age'] = current_year - df_planes['year']

# Unir la tabla de vuelos con la tabla de aviones para obtener la información de la edad del avión
df_flights_with_age = df_flights.merge(df_planes[['tailnum', 'age']], on='tailnum', how='left')

# Determinar la frecuencia de mantenimiento requerido para cada avión (por ejemplo, contando la cantidad de vuelos)
maintenance_frequency = df_flights_with_age.groupby('tailnum').size().reset_index(name='maintenance_count')

# Unir la frecuencia de mantenimiento con la información de la edad del avión
df_maintenance_age_relation = df_planes.merge(maintenance_frequency, on='tailnum', how='left')

# Analizar la relación entre la edad del avión y la frecuencia de mantenimiento
# Por ejemplo, podrías hacer un gráfico de dispersión o calcular la correlación

# Calcular la correlación entre la edad del avión y la frecuencia de mantenimiento requerido
correlation = df_maintenance_age_relation['age'].corr(df_maintenance_age_relation['maintenance_count'])
print("Correlación entre la edad del avión y la frecuencia de mantenimiento requerido:", correlation)
```

##### 9) ¿Cómo varía la puntualidad de salida de los vuelos para cada aerolínea a lo largo de los meses del año?

```python
import pandas as pd
import matplotlib.pyplot as plt

# Suponiendo que df_flights contiene los datos de los vuelos y df_airlines contiene los datos de las aerolíneas

# 1. Agrupar los vuelos por aerolínea y mes
grouped = df_flights.groupby(['carrier', 'month'])

# 2. Calcular el porcentaje de vuelos que salieron a tiempo para cada aerolínea en cada mes
on_time_percentage = grouped.apply(lambda x: (x['dep_delay'] <= 0).mean() * 100).reset_index(name='on_time_percentage')

# 3. Visualizar la puntualidad de salida de los vuelos para cada aerolínea a lo largo de los meses del año
for carrier in df_airlines['carrier']:
    data = on_time_percentage[on_time_percentage['carrier'] == carrier]
    plt.plot(data['month'], data['on_time_percentage'], label=carrier)

plt.title('Puntualidad de salida de los vuelos por aerolínea y mes')
plt.xlabel('Mes')
plt.ylabel('Porcentaje de vuelos a tiempo')
plt.legend(loc='upper left')
plt.xticks(range(1, 13))  # Establecer los meses en el eje x
plt.grid(True)
plt.show()
```

##### 1) ¿En que país y que avión se encuentra entre el 85% y el 70% de la cantidad de aterrisajes? Proporcione el nombre del país y el nombre del avión.

```python
import pandas as pd

# Suponiendo que el origen de los vuelos es igual al país donde se encuentra el aeropuerto
# Primero, necesitamos calcular la cantidad total de vuelos por avión y país
df_flights['total_flights'] = 1  # Agregar columna para contar los vuelos
df_flights['landing_country'] = df_flights['dest']  # Suponemos que el destino del vuelo es el país de aterrizaje

# Agrupar por avión y país y contar la cantidad total de vuelos
df_landings = df_flights.groupby(['tailnum', 'landing_country']).agg(total_flights=('total_flights', 'sum')).reset_index()

# Calcular el total de aterrizajes por avión
df_landings['total_landings'] = df_landings.groupby('tailnum')['total_flights'].transform('sum')

# Calcular el porcentaje de aterrizajes por avión y país
df_landings['landing_percentage'] = df_landings['total_flights'] / df_landings['total_landings'] * 100

# Filtrar los aviones con un porcentaje de aterrizajes entre el 85% y el 70%
filtered_landings = df_landings[(df_landings['landing_percentage'] <= 85) & (df_landings['landing_percentage'] >= 70)]

# Encontrar el país y avión con el porcentaje requerido
result = filtered_landings[['tailnum', 'landing_country']].head(1)

# Imprimir el resultado
print("El avión", result['tailnum'].values[0], "se encuentra en", result['landing_country'].values[0], "con entre el 85% y el 70% de aterrizajes.")
```

##### 2)¿Indique cual es el areopuerto con la temparatura más alta registrada en los datos?

```python
import pandas as pd

# Consulta SQL para encontrar el aeropuerto con la temperatura más alta
sql_query = """
SELECT origin, MAX(temp) AS max_temp
FROM weather
GROUP BY origin
ORDER BY max_temp DESC
LIMIT 1;
"""

# Ejecutar la consulta y leer los resultados en un DataFrame
df_max_temp = pd.read_sql(sql_query, mysql_driver)

# Imprimir el aeropuerto con la temperatura más alta
print("El aeropuerto con la temperatura más alta registrada es:", df_max_temp['origin'][0], "con una temperatura de", df_max_temp['max_temp'][0], "grados.")
```

c) Muestre una gráfica
i. de barras para la cantidad de las variables categoricas y discretas.
